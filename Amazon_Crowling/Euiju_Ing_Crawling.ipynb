# 성분 크롤링

```python
def amazon_crawling_Ing(soup, m):
    
    tmp_lst = []
    
    for x in range(m):
        #제품 성분
        try:
            tep = soup.select('#important-information > div:nth-child(3) > p:nth-child(3)')[x].text
            pattern = r'\([^)]*\)'
            prd_Ing = re.sub(pattern=pattern, repl='', string=tep)
        except IndexError:
            prd_Ing = None
            
        tmp_lst.append([prd_Ing])

    return tmp_lst
```

```python
driver = webdriver.Chrome('/Users/jeongeuiju/Amazon_Web_Crowling/chromedriver')

total_lst = []

for n in tqdm(range(4800)):
    url = df.Link[n]
    driver.get(url)
    time.sleep(1)
    
    
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    
    tmp_lst2 = amazon_crawling_Ing(soup, 1)
    total_lst.extend(tmp_lst2)
```

```python
df = pd.DataFrame(total_lst, columns = ['prd_Ing'])

df.to_csv('Crawling_Amazon_Ing1.csv', index_label = False)
```
